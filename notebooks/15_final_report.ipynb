**This notebook answers one research question.**

**Research Question:** What is the final summary of the whole project?

This notebook runs no new analysis. It loads the saved outputs, prints the verified headline
numbers, and these numbers back the final report and the executive summary.

## Load saved outputs and print the headline numbers

```python
import sys
from pathlib import Path
import pandas as pd

root = Path.cwd()
while not (root / "src").exists() and root != root.parent:
    root = root.parent
sys.path.insert(0, str(root))
from src import scoring

R = root / "version2" / "reports"

index = pd.read_csv(root / "results" / "competitiveness_index.csv", index_col=0)
clusters = pd.read_csv(R / "state_clusters.csv", index_col=0)["cluster"]
gap_nat = pd.read_csv(R / "gap_vs_national.csv", index_col=0)
gap_top = pd.read_csv(R / "gap_vs_top_group.csv", index_col=0)
priorities = pd.read_csv(R / "development_priorities.csv", index_col=0)
sweep = pd.read_csv(R / "scenario_sweep.csv", index_col=0)
ranked = index[index["Rank"].notna()]
FC = set(scoring.FACTOR_CONDITIONS)
ind_areas = {"Industry", "Manufacturing", "Investment", "MSMEs"}

print("states ranked:", len(ranked))
print("highest / lowest:", ranked["competitiveness_score"].idxmax(),
      round(ranked["competitiveness_score"].max(), 3), "/",
      ranked["competitiveness_score"].idxmin(), round(ranked["competitiveness_score"].min(), 3))
print("groups:", dict(clusters.value_counts()))
print("biggest weakness vs national -> basic:", int(gap_nat.idxmax(axis=1).isin(FC).sum()),
      "industry:", int((~gap_nat.idxmax(axis=1).isin(FC)).sum()))
print("biggest gap vs top group -> industry:", int((~gap_top.idxmax(axis=1).isin(FC)).sum()), "of 33")
print("industry-related main priority:", int(priorities["main_priority"].isin(ind_areas).sum()), "of 33")
print("scenario score gain:", f"{sweep['score_gain'].min():.3f} to {sweep['score_gain'].max():.3f}",
      "| rank move:", f"{sweep['rank_gain'].min()} to {sweep['rank_gain'].max()}")
```
