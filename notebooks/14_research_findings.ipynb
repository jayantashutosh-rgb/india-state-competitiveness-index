## Setup — load the saved outputs

```python
import sys
from pathlib import Path
import pandas as pd

root = Path.cwd()
while not (root / "src").exists() and root != root.parent:
    root = root.parent
sys.path.insert(0, str(root))
from src import scoring

R = root / "version2" / "reports"

index = pd.read_csv(root / "results" / "competitiveness_index.csv", index_col=0)
clusters = pd.read_csv(R / "state_clusters.csv", index_col=0)["cluster"]
gap_nat = pd.read_csv(R / "gap_vs_national.csv", index_col=0)
gap_top = pd.read_csv(R / "gap_vs_top_group.csv", index_col=0)
priorities = pd.read_csv(R / "development_priorities.csv", index_col=0)
sweep = pd.read_csv(R / "scenario_sweep.csv", index_col=0)
ranked = index[index["Rank"].notna()]

print("loaded all saved outputs. states:", len(ranked))
```

## Ranking recap (Notebook 07)

```python
print("states ranked:", len(ranked))
print("highest:", ranked["competitiveness_score"].idxmax(), round(ranked["competitiveness_score"].max(), 3))
print("lowest:", ranked["competitiveness_score"].idxmin(), round(ranked["competitiveness_score"].min(), 3))
```

## Groups recap (Notebook 10)

```python
for name, n in clusters.value_counts().items():
    print(f"{name}: {n}")
```

## Gaps recap (Notebook 11)

```python
FC = set(scoring.FACTOR_CONDITIONS)
cw = gap_nat.idxmax(axis=1)
dg = gap_top.idxmax(axis=1)
print("vs national average -> basic conditions:", int(cw.isin(FC).sum()), "| industry:", int((~cw.isin(FC)).sum()))
print("vs top group        -> basic conditions:", int(dg.isin(FC).sum()), "| industry:", int((~dg.isin(FC)).sum()))
```

## Priorities recap (Notebook 12)

```python
ind_areas = {"Industry", "Manufacturing", "Investment", "MSMEs"}
n = int(priorities["main_priority"].isin(ind_areas).sum())
print(f"industry-related main priority: {n} of {len(priorities)} states")
```

## Scenarios recap (Notebook 13)

```python
print(f"score gain range: {sweep['score_gain'].min():.3f} to {sweep['score_gain'].max():.3f}")
print(f"rank movement range: {sweep['rank_gain'].min()} to {sweep['rank_gain'].max()} places")
print(f"largest score gain: {sweep['score_gain'].idxmax()} (+{sweep['score_gain'].max():.3f})")
```
